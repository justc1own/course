```python
import numpy as np
import matplotlib.pyplot as plt

def f(x):
    return np.cos(x)

x_range = np.linspace(-0.5, 2, 400)
fig, ax = plt.subplots(figsize=(8, 8))

ax.plot(x_range, f(x_range), 'b', lw=2, label=r'$y=f(x)$')
ax.plot(x_range, x_range, 'k--', lw=1, label=r'$y=x$')
ax.text(0.1, 1.2, r'$f(x) = \cos(x)$', fontsize=16, color='blue')

x_n = 0.1
y_n = 0
ax.text(x_n, -0.07, '$x_0$', ha='center', va='top', fontsize=14, color='red')

for i in range(5):
    y_next = f(x_n)

    # стрелки
    ax.arrow(x_n, y_n, 0, y_next - y_n,
             head_width=0.04, head_length=0.06, fc='r', ec='r', length_includes_head=True)

    ax.arrow(x_n, y_next, y_next - x_n, 0,
             head_width=0.04, head_length=0.06, fc='r', ec='r', length_includes_head=True)

    x_n = y_next
    y_n = x_n

    if i < 3: # первые несколько точек подписаны
        ax.text(x_n, -0.07, f'$x_{i+1}$', ha='center', va='top', fontsize=14, color='red')

ax.set(xlabel="$x_n$", ylabel="$x_{n+1}$", title="Диаграмма Ламерея (спиральная сходимость)")
ax.grid(True, linestyle=':', alpha=0.7)
ax.legend(fontsize=12, loc='upper left')
ax.set_ylim(0, 1.5)
ax.set_xlim(0, 1.5)
ax.set_aspect('equal', adjustable='box')
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

def f(x, r):
    return r * x * (1 - x)

r = 2.9
x_0 = 0.11
n_iterations = 4

x_range = np.linspace(0, 1, 500)
fig, ax = plt.subplots(figsize=(9, 9))

ax.plot(x_range, f(x_range, r), 'b', lw=2, label=r'$y=f(x)$')
ax.plot(x_range, x_range, 'k--', lw=1, label=r'$y=x$')
ax.text(0.4, 1.05, fr'$f(x) = {r} \cdot x(1-x)$', fontsize=16, color='blue')

x_n = x_0
ax.text(x_n, -0.04, '$x_0$', ha='center', va='top', fontsize=14, color='darkgreen')

for i in range(n_iterations):
    y_next = f(x_n, r)

    # стрелки
    ax.arrow(x_n, x_n, 0, y_next - x_n,
             color='r', alpha=0.6, lw=0.8,
             head_width=0.02, head_length=0.03, length_includes_head=True)

    ax.arrow(x_n, y_next, y_next - x_n, 0,
             color='r', alpha=0.6, lw=0.8,
             head_width=0.02, head_length=0.03, length_includes_head=True)

    x_n = y_next

    if i < 3:
        ax.text(x_n, -0.07, f'$x_{i+1}$', ha='center', va='top', fontsize=14, color='red')


ax.set(xlabel="$x_n$", ylabel="$x_{n+1}$", title="Диаграмма Ламерея (хаотический режим)")
ax.grid(True, linestyle=':', alpha=0.7)
ax.legend(fontsize=12, loc='upper left')
ax.set_xlim(0, 1)
ax.set_ylim(0, 1.1)
ax.set_aspect('equal', adjustable='box')
plt.show()
```
