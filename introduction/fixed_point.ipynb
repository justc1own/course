```python
import numpy as np
import matplotlib.pyplot as plt

def f(x, r=2):
    return r * x * (1 - x)

fig, ax = plt.subplots(figsize=(9, 9))
ax.set_aspect('equal')

x_range = np.linspace(0, 1, 400)
ax.plot(x_range, f(x_range), 'b', lw=2, label=fr'$y = {2.0}x(1-x)$')
ax.plot(x_range, x_range, 'k--', lw=1, label=r'$y=x$')

x_n = 0.1
n_iterations = 6

for i in range(n_iterations):
    y_next = f(x_n)

    ax.plot([x_n, x_n], [x_n, y_next],
            color='r', alpha=0.8, lw=1.2)

    ax.plot([x_n, y_next], [y_next, y_next],
            color='r', alpha=0.8, lw=1.2)

    x_n = y_next

ax.annotate(fr'Неподвижная точка' + '\n' + r'$x^* = 0.5$',
             xy=(0.5, 0.5),
             xytext=(0.15, 0.55),
             arrowprops=dict(facecolor='black', shrink=0.05, width=1),
             fontsize=14,
             ha='center')

ax.set_title('Неподвижная точка', fontsize=16)
ax.set(xlabel="$x_n$", ylabel="$x_{n+1}$")
ax.grid(True, linestyle=':', alpha=0.7)
ax.legend(fontsize=12, loc='lower left')
ax.set_xlim(0, 1)
ax.set_ylim(0, 0.6)
plt.show()
```
